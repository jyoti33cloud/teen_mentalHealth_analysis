```python
# 1. IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# 2. LOAD DATA
df = pd.read_csv("Teen_Mental_Health_Dataset.csv")
```

```python
# Show basic info
print("First 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())
```

```python
# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Drop duplicates (if any)
df = df.drop_duplicates()
```

```python
# Age distribution
df['age'].plot(kind='hist', bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#  GENDER DISTRIBUTION

gender_counts = df['gender'].value_counts()

plt.figure()
plt.bar(gender_counts.index, gender_counts.values)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#platform usage
platform_counts = df['platform_usage'].value_counts()

plt.figure()
plt.bar(platform_counts.index, platform_counts.values)
plt.title("Platform Usage")
plt.xlabel("Platform")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#platform wise gender analysis
platform_gender = df.groupby(['platform_usage','gender']).size().unstack(fill_value=0)

print(platform_gender)

platform_gender.plot(kind='bar')
plt.title("Platform Usage by Gender")
plt.xlabel("Platform")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.tight_layout()
plt.show()
```

```python
#mental health overview
print(df[['stress_level','anxiety_level','addiction_level','sleep_hours']].describe())
```

```python
#stress distribution
plt.figure()
plt.hist(df['stress_level'], bins=10)
plt.title("Stress Level Distribution")
plt.xlabel("Stress Level")
plt.ylabel("Frequency")
plt.show()
```

```python
#social media vs stress
plt.figure()
plt.scatter(df['daily_social_media_hours'], df['stress_level'])
plt.title("Social Media vs Stress")
plt.xlabel("Social Media Hours")
plt.ylabel("Stress Level")
plt.show()
```

```python
#depression
dep_counts = df['depression_label'].value_counts()

plt.figure()
plt.bar(dep_counts.index.astype(str), dep_counts.values)
plt.title("Depression Distribution")
plt.xlabel("0 = No, 1 = Yes")
plt.ylabel("Count")
plt.show()
```
